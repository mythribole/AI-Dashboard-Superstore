```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/leonism/sample-superstore/master/data/superstore.csv"
df = pd.read_csv(url)
print(df.shape)
print(df.columns)
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
print(df.info)
```

```python
print(df.describe)
```

```python
df = df.dropna(subset=["Customer ID"])
print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df['Postal Code']=df['Postal Code'].fillna(df['Postal Code'].median())
```

```python
print(df.isnull().sum())
```

```python
print("Total sales: ",df['Sales'].sum())
```

```python
print("Total profit: ",df['Profit'].sum())
```

```python
print("Total Orders: ",len(df['Order ID'].unique()))
```

```python
print("Total customers: ",len(df['Customer ID'].unique()))
```

```python
print("Total products: ",len(df['Product ID'].unique()))
```

```python
region_sales=df.groupby('Region')['Sales'].sum()
plt.bar(region_sales.index, region_sales.values)
plt.title("Sales By Each Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
category=df.groupby('Category')['Profit'].sum()
plt.bar(category.index, category.values)
plt.title("Profit By Category")
plt.xlabel("category")
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Year'] = df['Order Date'].dt.year
yearly_sales = df.groupby('Order Year')['Sales'].sum()
plt.plot(yearly_sales.index, yearly_sales.values)
plt.title("Year-wise Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()
plt.bar(segment_sales.index, segment_sales.values)
plt.title("Sales by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.bar(top_products.index, top_products.values)
plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```
